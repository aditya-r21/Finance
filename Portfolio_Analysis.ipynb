## Function to Plot Data

```python
def plot_data(data: pd.DataFrame, title: str, label: str):
    plt.figure(figsize=(12, 4))
    plt.plot(data['Close'], label=label)
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_data2(data: pd.DataFrame, title: str, label: str):
    plt.figure(figsize=(12, 4))
    plt.plot(data.index, data.values, label = label)
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

Apple has a beta of 1.18, which tells that stocks of Apple are 18% more volatile than the broader market. This data could be considered good for risk-tolerant investors expecting potential high returns. But, on the other hand, this number may suggest that the stock price is too volatile to invest for conservative investors trying to make risk-free investments.

### **Technical Indicators**

#### **Moving Average**

A moving average (MA) is a technical indicator that calculates the average price of an asset over a specific period of time. It could be used to interpret and analyze the trend of the stock price of a company.

```python
short_ma = aapl_daily_data['Close'].rolling(window=20).mean()
long_ma = aapl_daily_data['Close'].rolling(window=50).mean()
plot_data2(short_ma, "Apple 20-day Moving Average", "Moving Average")
plot_data2(long_ma, "Apple 50-day Moving Average", "Moving Average")
```

#### **RSI (Relative Strength Index)**

RSI is a momentum oscillator used to measure the speed and change of price movements. It helps investors and traders assess whether a stock or other asset is overbought or oversold.

* **Above 70:** The asset is considered overbought, suggesting it may be overvalued and could be due for a pullback or correction.
* **Below 30:** The asset is considered oversold, indicating it may be undervalued and could be due for a rebound or upward movement.

```python
delta = aapl_daily_data['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
rsi = 100 - (100 / (1 + rs))
plot_data2(rsi, "Apple RSI", "RSI")
```

#### **MACD (Moving Average Convergence Divergence)**

MACD is used to identify potential buy and sell signals, as well as the strength and direction of a stock's trend.

* **Bullish Crossover:** When the MACD line crosses above the signal line, it’s a potential buy signal, indicating that upward momentum may be increasing.
* **Bearish Crossover:** When the MACD line crosses below the signal line, it’s a potential sell signal, indicating that downward momentum may be increasing.

```python
short_ema = aapl_daily_data['Close'].ewm(span=12, adjust=False).mean()
long_ema = aapl_daily_data['Close'].ewm(span=26, adjust=False).mean()
macd = short_ema - long_ema
signal = macd.ewm(span=9, adjust=False).mean()
histogram = macd - signal
plt.figure(figsize=(12, 4))
plt.plot(macd, label='MACD')
plt.plot(signal, label='Signal')
plt.title('Apple MACD v/s Signal')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
plot_data2(histogram, "Apple Histogram", "Histogram")
```

## **Ratio Analysis**

```python
stock = yf.Ticker("AAPL")
income_statement = stock.financials.transpose()
balance_sheet = stock.balance_sheet.transpose()
cash_flow = stock.cashflow.transpose()
stock_data = stock.history(period="1y")
```

### **Valuation Ratios**

#### **P/E (Price-to-Earnings Ratio)**

It compares the company's current share price to its earnings per share (EPS), giving insight into how much investors are willing to pay for each dollar of earnings. It helps investors evaluate the relative value of a company's stock.

1. **High P/E Ratio:**
* *Possible Growth:* A high P/E ratio may indicate that investors expect higher future growth and are willing to pay a premium for the stock.
* *Overvaluation:* Conversely, a high P/E could also suggest that the stock is overvalued compared to its earnings.

2. **Low P/E Ratio:**
* *Undervaluation:* A low P/E may indicate that the stock is undervalued or that the company is experiencing difficulties.

* *Value Stock:* Some investors see low P/E stocks as potential bargains, especially if they believe the company's earnings will improve.

```python
pe_ratio = stock_data['Close'][-1] / income_statement['Net Income'].iloc[0]
print(f"P/E Ratio: {pe_ratio} \n")
```

#### **P/B (Price-to-Book Ratio)**

It provides insight into how much investors are willing to pay for each dollar of a company’s net assets.

1. **P/B < 1:**
* *Undervalued Stock:* The market price is lower than the company's book value. This might suggest a potential buying opportunity.
* *Distressed Company:* It could indicate that the market expects the company’s assets to be worth less than their book value, possibly due to financial trouble or poor future prospects.

2. **P/B > 1:**
* *Overvalued Stock:* The market price is higher than the company's book value. Investors might be paying a premium for perceived growth or other factors.
* *Strong Company:* It could also indicate that investors believe the company has strong growth potential, high earnings power, or valuable intangible assets not fully captured in the book value.


```python
pb_ratio = stock_data['Close'][0] / balance_sheet['Stockholders Equity'].iloc[0]
print(f"P/B Ratio: {pb_ratio} \n")
```

#### **EV/EBITDA (Enterprise Value to Earnings Before Interest, Taxes, Depreciation, and Amortization Ratio)**

It compares a company's enterprise value (EV) to its EBITDA, providing a more comprehensive view of valuation that considers the entire capital structure, rather than just equity. It is used to assess the value of a company.

```python
ev = (stock_data['Close'][-1] * stock.info['sharesOutstanding']) + balance_sheet['Total Liabilities Net Minority Interest'].iloc[0] - cash_flow['Cash Flow From Continuing Operating Activities'].iloc[0]
ebitda = income_statement['EBITDA'].iloc[0]
ev_ebitda = ev / ebitda
print(f"ev: {ev}")
print(f"ebitda: {ebitda}")
print(f"ev/ebitda: {ev_ebitda} \n")
```

### **Profitability Ratios**

### **Working Capital Management**

#### **Current Ratio**

It measures a company’s ability to pay off its short-term liabilities with its short-term assets.

1. **Current Ratio > 1:**
* *Healthy Liquidity:* The company has more current assets than current liabilities, suggesting it is in a good position to cover its short-term obligations.
* *Potential Efficiency Issues:* The company is not using its assets efficiently or is hoarding cash rather than investing it back into the business.

2. **Current Ratio < 1:**
* *Liquidity Risk:* The company may not have enough assets to cover its short-term liabilities, which could indicate potential liquidity problems.
* *Potential Distress:* his situation may raise concerns about the company's ability to meet its obligations as they come due, potentially leading to financial distress if not addressed.

```python
current_ratio = balance_sheet['Current Assets'].iloc[0] / balance_sheet['Current Liabilities'].iloc[0]
print(f"Current Ratio: {current_ratio}")
```

#### **Quick Ratio**

It is used to evaluate a company's ability to meet its short-term liabilities with its most liquid assets. It is a more stringent measure of liquidity than the current ratio because it excludes inventory and other less liquid assets from current assets.

```python
quick_ratio = (balance_sheet['Current Assets'].iloc[0] - balance_sheet['Inventory'].iloc[0]) / balance_sheet['Current Liabilities'].iloc[0]
print(f"Quick Ratio: {quick_ratio}")
```

# **Portfolio Analysis**

## **Portfolio Construction & Evaluation**

```python
import datetime as dt
start_date = dt.datetime.now() - dt.timedelta(days=3*365)
end_date = dt.datetime.now()
data = yf.download(['AAPL', 'AMZN', 'ACN', 'ADBE', 'AMD', '^GSPC'], start = start_date, end = end_date)['Close']
plt.figure(figsize=(12, 4))
plt.plot(data[['AAPL', 'AMZN', 'ACN', 'ADBE', 'AMD']].index, data[['AAPL', 'AMZN', 'ACN', 'ADBE', 'AMD']].values, label=['Apple', 'Amazon', 'Accenture', 'Adobe', 'AMD'])
plt.title("Daily Adjusted Close Price")
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
apple_weight = 0.30
remaining_weight = 0.70
remaining_prices = data[['AMZN', 'ACN', 'ADBE', 'AMD']]
price_weights = remaining_prices.iloc[-1] / remaining_prices.iloc[-1].sum()
weights = price_weights * remaining_weight
weights['AAPL'] = apple_weight
weights = weights / weights.sum()
weights
plt.pie(weights, labels=weights.index, autopct='%1.1f%%')
plt.title('Portfolio Weights')
plt.show()
```

```python
returns = data.pct_change().dropna()
portfolio_price = (data[['AAPL', 'AMZN', 'ACN', 'ADBE', 'AMD']] * weights).sum(axis=1)
portfolio_returns = (returns[['AAPL', 'AMZN', 'ACN', 'ADBE', 'AMD']] * weights).sum(axis=1)
plot_data2(portfolio_price, "Portfolio Price", "Portfolio Price")
plot_data2(portfolio_returns, "Portfolio Returns", "Portfolio Returns")
```

### **Portfolio Metrics**

#### **Sharpe Ratio**

```python
risk_free_rate = 0.02     # Annual risk free rate takes as 2%
excess_returns = portfolio_returns - risk_free_rate / 252
sharpe = np.sqrt(252) * excess_returns.mean() / excess_returns.std()
print(f"Sharpe Ratio: {sharpe}")
```

#### **Treynor Ratio**

```python
benchmark_returns = returns['^GSPC']
beta = np.cov(portfolio_returns, benchmark_returns)[0, 1] / np.var(benchmark_returns)
excess_returns = portfolio_returns.mean() - risk_free_rate / 252
treynor = np.sqrt(252) * excess_returns / beta
print(f"Treynor Ratio: {treynor}")
```

#### **Jensen's Alpha**

```python
import statsmodels.api as sm
X = sm.add_constant(benchmark_returns - risk_free_rate / 252)
y = portfolio_returns - risk_free_rate / 252
model = sm.OLS(y, X).fit()
jensen_alpha = model.params[0] * 252  # Annualize alpha
jensen_beta = model.params[1]
print(f"Jensen Alpha: {jensen_alpha}")
print(f"Jensen Beta: {jensen_beta}\n")
```

### **Performance Comparison**

```python
cumulative_portfolio_returns = (1 + portfolio_returns).cumprod() - 1
cumulative_benchmark_returns = (1 + benchmark_returns).cumprod() - 1
plt.figure(figsize=(12, 4))
plt.plot(cumulative_portfolio_returns, label='Portfolio')
plt.plot(cumulative_benchmark_returns, label='Benchmark (S&P 500)')
plt.title('Portfolio vs Benchmark Performance')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.legend()
plt.grid(True)
plt.show()
```

The deviations between Portfolio returns and Benchmark index returns are because:
*  A portfolio might be more concentrated in certain sectors or asset classes, leading to different performance outcomes.
*  Differences in investment choices, strategies, and timing decisions can cause deviations from benchmark performance.
*  Portfolios may also incur operational expenses that impact returns, whereas benchmark indices typically do not include such costs.
*  Economic conditions, market trends, and geopolitical events can affect portfolio and benchmark performance differently.
*  Currency fluctuations can impact returns compared to a benchmark that may be less exposed to foreign currencies.



```python
yearly_performance = lambda x: (1 + x).resample('Y').prod() - 1
portfolio_yearly_performance = yearly_performance(portfolio_returns)
benchmark_yearly_performance = yearly_performance(benchmark_returns)
performance_comparison = pd.DataFrame({'Portfolio': portfolio_yearly_performance,'Benchmark': benchmark_yearly_performance})
plt.figure(figsize=(12, 4))
plt.plot(portfolio_yearly_performance, label='Portfolio')
plt.plot(benchmark_yearly_performance, label='Benchmark (S&P 500)')
plt.title('Yearly Portfolio vs Benchmark Performance')
plt.xlabel('Date')
plt.ylabel('Yearly Returns')
plt.legend()
plt.grid(True)
plt.show()
```

## **Single Stock vs. Portfolio Comparison**

### **Risk Reward Comparison**

#### **Annualized Return**

```python
apple_returns = returns['AAPL']

def annualized_return(daily_returns):
    compounded_growth = (1 + daily_returns).prod()
    n_periods = daily_returns.shape[0]
    return compounded_growth ** (252 / n_periods) - 1

apple_annual_return = annualized_return(apple_returns)
portfolio_annual_return = annualized_return(portfolio_returns)
print(f"Apple Annualized Return: {apple_annual_return:.2%}")
print(f"Portfolio Annualized Return: {portfolio_annual_return:.2%}")
```

#### **Annualized Volatility**

```python
def annualized_volatility(daily_returns):
    return daily_returns.std() * np.sqrt(252)

apple_annual_volatility = annualized_volatility(apple_returns)
portfolio_annual_volatility = annualized_volatility(portfolio_returns)
print(f"Apple Annualized Volatility: {apple_annual_volatility:.2%}")
print(f"Portfolio Annualized Volatility: {portfolio_annual_volatility:.2%}")
```

#### **Sharpe Ratio**

```python
def sharpe_ratio(daily_returns, risk_free_rate=0.01):
    excess_returns = daily_returns - risk_free_rate / 252
    return np.sqrt(252) * excess_returns.mean() / excess_returns.std()

apple_sharpe_ratio = sharpe_ratio(apple_returns)
portfolio_sharpe_ratio = sharpe_ratio(portfolio_returns)
print(f"Apple Sharpe Ratio: {apple_sharpe_ratio:.2f}")
print(f"Portfolio Sharpe Ratio: {portfolio_sharpe_ratio:.2f}")
```

## **Correlation-Based Portfolio**

```python
correlation_matrix = returns[['AAPL', 'AMZN', 'ACN', 'ADBE', 'AMD']].corr()
correlation_matrix
```

```python
correlations = correlation_matrix['AAPL'].drop('AAPL')
inv_correlations = 1 - correlations # Invert correlations (lower correlation gets higher weight)
remaining_prices = data.iloc[-1].drop('AAPL') # Changed from returns.iloc[-1]
price_weights = remaining_prices / remaining_prices.sum()
combined_weights = inv_correlations * price_weights
combined_weights = (combined_weights / combined_weights.sum()) * (1 - 0.3)
combined_weights["AAPL"] = 0.3
combined_weights = combined_weights[['AAPL','AMZN','ACN','ADBE','AMD']]
plt.pie(combined_weights, labels=combined_weights.index, autopct='%1.1f%%')
plt.title('Correlation-Based Weights')
plt.show()
```

```python
correlation_returns = (returns[['AAPL', 'AMZN', 'ACN', 'ADBE', 'AMD']] * combined_weights).sum(axis=1)
cumulative_price_weighted = (1 + portfolio_returns).cumprod() - 1
cumulative_correlation_based = (1 + correlation_returns).cumprod() - 1
performance_comparison = pd.DataFrame({'Price-Weighted Portfolio': cumulative_price_weighted,'Correlation-Based Portfolio': cumulative_correlation_based})
plt.figure(figsize=(12, 4))
plt.plot(performance_comparison['Price-Weighted Portfolio'], label='Price-Weighted Portfolio')
plt.plot(cumulative_benchmark_returns, label='Benchmark (S&P 500)')
plt.plot(performance_comparison['Correlation-Based Portfolio'], label='Correlation-Based Portfolio')
plt.title('Portfolio Performance Comparison')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.legend()
plt.grid(True)
plt.show()
```

Here, it could be clearly observed that Correlation-Based Portfolio outperforms Price-Weighted Portfolio. The reasons may be:
*  By selecting assets with low or negative correlations, the correlation-based portfolio reduces the risk that all investments will decline simultaneously.
*  A price-weighted portfolio may have a higher concentration in assets with higher prices, regardless of their underlying fundamentals or market risk. This can lead to a higher risk if those high-priced assets perform poorly.
*  By focusing on a balanced mix of assets with varying correlations, a correlation-based portfolio can capture different sources of return and maintain stability, which is beneficial for long-term growth.
*  The allocation of capital is based on minimizing correlations rather than just prices. This can lead to more efficient use of capital and better risk-adjusted returns. Selecting assets based on correlation can lead to more thoughtful inclusion of assets that provide better risk-adjusted returns.
*  Capital allocation based on the price of assets rather than their risk or correlation can lead to suboptimal allocation and potentially lower returns if high-priced assets do not perform well.
